  .Determine the top three most common cuisines in Dataset  

```python
#find value count of cuisines column

a = df['Cuisines'].value_counts()
a
```

#### Calculate the percentage of resturants that serve each of the top cuisines

```python
#length of the dataset
tot_res = len(df)
tot_res
```

```python
per =(top_three.values/tot_res)*100
```

```python
data = dict(zip(top_three.index,per))
data
```

```python
# create the dataframe with their percentage
a = pd.DataFrame(data.items(),columns =['Cuisines','percentage'])
a
```

```python
# pie chart plot
plt.title('percentage of restaurants that serve each of the top cuisines')
plt.pie(a['percentage'],labels= a['Cuisines'],autopct='%0.2f%%',explode=[0.05,0.05,0.05],startangle=90)
plt.show()
```
